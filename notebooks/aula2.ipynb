### Desafio 3 - Preço do metro quadrado por bairro e plotar em um gráfico ideal

#### Criando coluna `Valor_m2`

```python
dados_vendas['Valor_m2'] = dados_vendas['Valor_anuncio_float']/dados_vendas['Metragem']
```

```python
dados_bairro = dados_vendas.groupby('Bairro').mean().round(2)
```

```python
dados_bairro.reset_index(inplace=True)
dados_bairro.head()
```

#### Bairros com mais imóveis à venda

```python
top_bairros = dados_vendas['Bairro'].value_counts()[:10].index
top_bairros
```

```python
dados_bairro.query('Bairro in @top_bairros')
```

### Desafio 4 - Explorar as bibliotecas de visualizações e colocar as suas conclusões

```python
ax = sns.barplot(x='Bairro', y='Valor_m2', data=dados_bairro.query('Bairro in @top_bairros'))
plt.title('\nBairros com mais imóveis à venda\n')
ax.tick_params(axis='x', rotation=45)
```

```python
plt.figure(figsize=(15,8))
ax = sns.boxplot(data=dados_vendas.query('Bairro in @top_bairros'), x='Bairro', y='Valor_anuncio_float')
ax.tick_params(axis='x', rotation=45)
```

```python
plt.figure(figsize=(15,8))
ax = sns.boxplot(data=dados_vendas.query('Bairro in @top_bairros'), x='Bairro', y='Valor_m2')
ax.tick_params(axis='x', rotation=45)
```
